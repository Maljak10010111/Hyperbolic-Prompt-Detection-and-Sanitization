```python

import os
import csv
# read a csv asd a datafra,e
import pandas as pd

csv_path = '/mnt/ssd1/mary/Diffusion-Models-Embedding-Space-Defense/.cache/mma-diffusion-nsfw-adv-prompts.csv'
df = pd.read_csv(csv_path)
```

```python
print(df.head())
```

```python
# get only clean prompts
clean_df = df['clean_prompt']
```

```python
# print first 5 clean prompts
print(clean_df.head())
```

```python
# save it to a csv file

import csv
import random

random.seed(42)
output_file = "mma_clean_prompts.csv"

with open(output_file, "w", newline='') as f:
    writer = csv.writer(f, quoting=csv.QUOTE_MINIMAL)
    writer.writerow(["case_number", "prompt", "categories", "evaluation_seed"])
    for i, example in enumerate(clean_df):
        case_number = i
        prompt = example  # Let the csv module handle quoting
        categories = "SAFE"
        evaluation_seed = random.randint(0, 32_000)
        writer.writerow([case_number, prompt, categories, evaluation_seed])

print(f"Text prompts saved to {output_file}")
```
