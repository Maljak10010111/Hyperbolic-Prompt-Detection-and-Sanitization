```python
from datasets import load_dataset
import os



ds = load_dataset("aimagelab/ViSU-Text")

print(ds)
```

```python
# get the test split and have access to the text prompts
test_ds = ds["test"]
print(test_ds)
```

```python
import csv
import random

random.seed(42)
output_file = "visu_text_prompts.csv"

with open(output_file, "w", newline='') as f:
    writer = csv.writer(f, quoting=csv.QUOTE_MINIMAL)
    writer.writerow(["case_number", "prompt", "categories", "evaluation_seed"])
    for i, example in enumerate(test_ds):
        case_number = i
        prompt = example["nsfw"]  # Let the csv module handle quoting
        categories = "NSFW"
        evaluation_seed = random.randint(0, 32_000)
        writer.writerow([case_number, prompt, categories, evaluation_seed])

print(f"Text prompts saved to {output_file}")
```
